```python
#dfextotal.loc[dfextotal['ปี']==2019,'มูลค่า (บาท)'].sum()
dfsumy={'2019':[dfextotal.loc[dfextotal['ปี']==2019,'มูลค่า (บาท)'].sum()],'2020':[dfextotal.loc[dfextotal['ปี']==2020,'มูลค่า (บาท)'].sum()],'2021':[dfextotal.loc[dfextotal['ปี']==2021,'มูลค่า (บาท)'].sum()]}
dfsumy=pd.DataFrame(data=dfsumy)
#dfsumy
#plt.style.use('seaborn')
#dfsumy.plot.bar()
```

```python
setmonth=list(set(dfextotal['เดือน ปี']))
setmonth=sorted(setmonth)

for i in range(len(setmonth)):
    setmonth[i]=setmonth[i].strftime("%Y/%m/%d")
    
#print(setmonth)

mnthrev=[0]*len(setmonth)

#mnthrev[1]=mnthrev[1]+dfextotal.loc[dfextotal['เดือน ปี']==setmonth[1],'มูลค่า (บาท)'].sum()

for i in range(len(setmonth)):
    mnthrev[i]=round(((mnthrev[i]+dfextotal.loc[dfextotal['เดือน ปี']==setmonth[i],'มูลค่า (บาท)'].sum())/1000000),2)
    
#print(mnthrev)
```

```python
dfsummnth={'เดือน ปี':setmonth,'มูลค่า (ล้านบาท)':mnthrev}
dfsummnth=pd.DataFrame(data=dfsummnth)

plt.figure(figsize=(15,5),dpi=150)
plt.suptitle('=== Monthly Export Revenue ===')
plt.plot(dfsummnth['เดือน ปี'],dfsummnth['มูลค่า (ล้านบาท)'],label="Export Revenue")
plt.legend()
plt.xticks(rotation=90)
```

```python
HS4dg21=sorted(list(set(dfextotal['HS 4dg'])))

sumhs21=[0]*len(HS4dg21)

for i in range(len(HS4dg21)):
    #sumhs21[i]=round(((sumhs21[i]+dfextotal.loc[dfextotal['HS 4dg']==HS4dg21[i],'มูลค่า (บาท)'].sum())/1000000),2)
    sumhs21[i]=round(((sumhs21[i]+dfex64.loc[dfex64['HS 4dg']==HS4dg21[i],'มูลค่า (บาท)'].sum())/1000000),2)


dfaccusum21={'HS4dg':HS4dg21,'มูลค่า 21 (ล้านบาท)':sumhs21}
dfaccusum21=pd.DataFrame(data=dfaccusum21)
#dfaccusum21
```

```python
HS4dg19=sorted(list(set(dfex62['HS 4dg'])))


sumhs19=[0]*len(HS4dg19)

for i in range(len(HS4dg19)):
    sumhs19[i]=round(((sumhs19[i]+dfextotal.loc[(dfextotal['HS 4dg']==HS4dg19[i])&(dfextotal['เดือน ปี']=='2019-01-01'),'มูลค่า (บาท)'].sum())/1000000),2)

dfaccusum19={'HS4dg':HS4dg19,'มูลค่า 19 (ล้านบาท)':sumhs19}
dfaccusum19=pd.DataFrame(data=dfaccusum19)
dfaccusum19
```

```python
dfaccusum=dfaccusum19.merge(dfaccusum21, how='outer').fillna(0)

dfaccusum['Diff 19-21 (ล้านบาท)']=dfaccusum['มูลค่า 21 (ล้านบาท)']-dfaccusum['มูลค่า 19 (ล้านบาท)']
dfaccusum['Diff 19-21 (%)']=round(((dfaccusum['มูลค่า 21 (ล้านบาท)']/dfaccusum['มูลค่า 19 (ล้านบาท)'])-1)*100,2)
top5=dfaccusum.sort_values(by=['Diff 19-21 (ล้านบาท)','Diff 19-21 (%)'],ascending = [False, True]).head()

gwroth=dfaccusum['มูลค่า 21 (ล้านบาท)'].sum()-dfaccusum['มูลค่า 19 (ล้านบาท)'].sum()
pgwroth=round(((dfaccusum['มูลค่า 21 (ล้านบาท)'].sum()/dfaccusum['มูลค่า 19 (ล้านบาท)'].sum())-1)*100,2)
#pgwroth
#print(top5)

dfaccusum=dfaccusum.fillna(0)
#dfaccusum.info()
top5g=dfaccusum.loc[dfaccusum['มูลค่า 19 (ล้านบาท)']>0.1].sort_values(by=['Diff 19-21 (%)','Diff 19-21 (ล้านบาท)'],ascending = [False, False]).head()
top5g
```

```python
country19=sorted(list(set(dfex62['ประเทศ'])))
sumc19=[0]*len(country19)

for i in range(len(country19)):
    sumc19[i]=round(((sumc19[i]+dfex62.loc[dfex62['ประเทศ']==country19[i],'มูลค่า (บาท)'].sum())/1000000),2)

countrysum19={'ประเทศ':country19,'มูลค่า 19 (ล้านบาท)':sumc19}
countrysum19=pd.DataFrame(data=countrysum19)
#countrysum19


country21=sorted(list(set(dfextotal['ประเทศ'])))
sumc21=[0]*len(country21)

for i in range(len(country21)):
    sumc21[i]=round(((sumc21[i]+dfextotal.loc[dfextotal['ประเทศ']==country21[i],'มูลค่า (บาท)'].sum())/1000000),2)

countrysum21={'ประเทศ':country21,'มูลค่า 21 (ล้านบาท)':sumc21}
countrysum21=pd.DataFrame(data=countrysum21)

#countrysum21
countrysum=countrysum19.merge(countrysum21, how='outer').fillna(0)
countrysum['Diff 19-21 (ล้านบาท)']=countrysum['มูลค่า 21 (ล้านบาท)']-countrysum['มูลค่า 19 (ล้านบาท)']
top5country=countrysum.sort_values(by=['Diff 19-21 (ล้านบาท)'],ascending = [False]).head()
#top5country
```

```python
plt.figure(figsize=(15,5),dpi=150)
plt.suptitle('=== Top5 Country Export Revenue ===')
plt.bar(top5country['ประเทศ'],top5country['Diff 19-21 (ล้านบาท)'])
```

```python
dfsummnth['มูลค่าสะสม (ล้านบาท)']=dfsummnth['มูลค่า (ล้านบาท)'].cumsum()

#print(dfsummnth)
```

```python
accu=list(dfsummnth['มูลค่าสะสม (ล้านบาท)'])

per=[0]*len(accu)

per[0]=0
for i in range(len(accu)-1):
    per[i+1]=round(((accu[i+1]/accu[i])-1)*100,2)

dfsummnth['%growth (ล้านบาท)']=per
dfsummnth['%growth (ล้านบาท)']=dfsummnth['%growth (ล้านบาท)'].cumsum()

############################################################################################################

top1=[0]*len(setmonth)
for i in range(len(setmonth)):
    top1[i]=round((dfextotal.loc[(dfextotal['HS 4dg']==top5.iloc[0,0])&(dfextotal['เดือน ปี']==dfsummnth.iloc[i,0]),'มูลค่า (บาท)'].sum())/1000000,2)
    
dfsummnth['Top-1']=top1
dfsummnth['Top-1 Accu']=dfsummnth['Top-1'].cumsum()

pertop1=[0]*len(accu)


pertop1[0]=0
for i in range(len(accu)-1):
    pertop1[i+1]=round(((dfsummnth.iloc[i+1,5]/dfsummnth.iloc[i,5])-1)*100,2)

dfsummnth['Top-1 %Growth']=pertop1
dfsummnth['Top-1 Acc %Growth']=dfsummnth['Top-1 %Growth'].cumsum()
#dfsummnth

############################################################################################################

top1g=[0]*len(setmonth)
for i in range(len(setmonth)):
    top1g[i]=round((dfextotal.loc[(dfextotal['HS 4dg']==top5g.iloc[0,0])&(dfextotal['เดือน ปี']==dfsummnth.iloc[i,0]),'มูลค่า (บาท)'].sum())/1000000,2)
    
dfsummnth['Top-1 Growth']=top1g
dfsummnth['Top-1 Growth Accu']=dfsummnth['Top-1 Growth'].cumsum()

pertop1g=[0]*len(accu)

pertop1g[0]=0
for i in range(len(accu)-1):
    pertop1g[i+1]=round(((dfsummnth.iloc[i+1,9]/dfsummnth.iloc[i,9])-1)*100,2)

dfsummnth['Top-1 Growth %Growth']=pertop1g
dfsummnth['Top-1 Growth Acc %Growth']=dfsummnth['Top-1 Growth %Growth'].cumsum()

############################################################################################################

plt.figure(figsize=(15,5),dpi=150)
plt.suptitle('=== Monthly Accu %Growth ===')
plt.plot(dfsummnth['เดือน ปี'],dfsummnth['%growth (ล้านบาท)'],label = "Overall Acc %Growth")
plt.plot(dfsummnth['เดือน ปี'],dfsummnth['Top-1 Acc %Growth'],color='red',label = "Top-1 Rev Growth")
plt.plot(dfsummnth['เดือน ปี'],dfsummnth['Top-1 Growth Acc %Growth'],color='orange',label = "Top-1 % Growth")
plt.legend()
plt.xticks(rotation=90)

#dfsummnth
```
